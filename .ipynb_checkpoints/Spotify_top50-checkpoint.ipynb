```python
import chartify
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('spotify.csv',encoding='latin-1')
```

```python
df.head()
```

```python
df.sort_values("Popularity", axis = 0, ascending = False, inplace = True, na_position ='last')
df.head()
```

```python
top_df = df[0:5]
top_df
```

```python
# create plot
fig, ax = plt.subplots()
index = np.arange(50)
bar_width = 0.35
opacity = 0.8
plt.figure(figsize=(12,9))

rects1 = plt.bar(index, df['Energy'], bar_width, alpha=opacity, color='b', label='Energy')

rects2 = plt.bar(index + bar_width, df['Danceability'], bar_width, alpha=opacity, color='g', label='Danceability')

plt.xlabel('Songs Attribute')
plt.ylabel('Values')
plt.title('Music')
plt.xticks(index + bar_width, df['Track_Name'])
plt.legend()

plt.tight_layout()
plt.show()
```

```python
print("Mean value for danceability:", df['Danceability'].mean())
sns.distplot(df['Danceability'])
plt.show()
print("Mean value for energy:", df['Energy'].mean())
sns.distplot(df['Energy'])
plt.show()
```

```python
numeric = df.drop(['Track_Name','Artist_Name', 'Genre'], axis=1)
small = numeric.drop(['Length','Loudness_dB','Beats_Per_Minute'], axis=1)
sns.set_palette('pastel')
small.mean().plot.bar()
plt.title('Mean Values of Audio Features')
plt.show()
```
